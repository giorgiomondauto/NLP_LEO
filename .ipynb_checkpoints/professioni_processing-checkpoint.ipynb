```python
import re
import pandas as pd
import nltk
import matplotlib
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('words')
import pickle
from nltk import word_tokenize
from nltk.corpus import stopwords
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
```

```python
professioni = pd.read_csv('professioni.csv')
professioni.head(10)
```

```python
professioni.Descrizione_professione = professioni.Descrizione_professione.apply(lambda x: x.lower().split('\n'))
professioni.Descrizione_professione = professioni.Descrizione_professione.apply(lambda x: [i.split("'") for i in x])

professioni.Descrizione_professione2 = professioni.Descrizione_professione2.apply(lambda x: x.lower().split('\n'))
professioni.Descrizione_professione2 = professioni.Descrizione_professione2.apply(lambda x: [i.split("'") for i in x])

professioni.Conoscenze = professioni.Conoscenze.apply(lambda x: str(x).lower().split('\n'))
professioni.Conoscenze = professioni.Conoscenze.apply(lambda x: [i.split("'") for i in x])
professioni['Descrizione_all'] = professioni.Descrizione_professione + professioni.Descrizione_professione2 + professioni.Conoscenze
professioni['Descrizione_all'] = professioni.Descrizione_all .apply(lambda x: [ i for t in x for i in t])
professioni['Descrizione_all']= professioni['Descrizione_all'].apply(lambda x: [i.split("/") for i in x])
professioni['Descrizione_all'] = professioni.Descrizione_all .apply(lambda x: [ i for t in x for i in t])

professioni['Descrizione_all'] = professioni.Descrizione_all .apply(lambda x: list(set(x)))
```

```python
# to remove empty space from the subgroup list of values
for i in range(0,professioni.Descrizione_all.shape[0]):
    while("" in professioni.Descrizione_all.iloc[i]) : 
        professioni.Descrizione_all.iloc[i].remove("") 
```

```python
# remove punctuation and tokenize 
# professioni.Descrizione_all= professioni.Descrizione_all.apply(lambda x: word_\tokenize(x)) # lowercase each word
professioni.Descrizione_all = professioni.Descrizione_all.apply(lambda x: list(set([t for i in x for t in word_tokenize(i)])))
professioni.Descrizione_all
```

```python
professioni.Conoscenze = professioni.Conoscenze.apply(lambda x: [i for t in x for i in t])
professioni
```

```python
Conoscenze_lista = []
for i in professioni.Conoscenze:
    for item in i:
        if item != 'nan':
            Conoscenze_lista.append(item)
Conoscenze_lista
```

```python
# remove words from description with length < 4 if the word is not part of 'Conoscenze_lista'
professioni.Descrizione_all = professioni.Descrizione_all.apply(lambda x: [i for i in x if (len(i) > 3 or i in Conoscenze_lista)])
professioni.Descrizione_all.head() # note the word 'vimercate' in ro
```

```python
## remove words like dell, della, dello, sull, degl. etc.
professioni.Descrizione_all = professioni.Descrizione_all.apply(lambda x: [i for i in x if not (i.startswith('nan') or\
                                                                    i.startswith('dell') \
                                                                    or i.startswith('sull')\
                                                                     or i.startswith('degl')
                                                                     or i.startswith('dall')
                                                                     or i.startswith('all'))])
```

```python
column = ['Conoscenze','Descrizione_all','Group']
info_data = professioni[column]
info_data.Conoscenze = info_data.Conoscenze.apply(lambda x: ' '.join(x))
info_data.Descrizione_all = info_data.Descrizione_all.apply(lambda x: ' '.join(x))
info_data['info'] = info_data['Descrizione_all'] + ' ' + info_data['Conoscenze']
info_data['info'] = info_data['info'].apply(lambda x: [i for i in word_tokenize(x)])
info_data.drop(['Conoscenze','Descrizione_all'], axis = 1, inplace = True)
info_data.head()
```

```python
info_data['info'] = info_data['info'].apply(lambda x: [i for i in x if (len(i) > 3 or i in Conoscenze_lista)])

info_data['info']= info_data['info'].apply(lambda x: [i for i in x if not (i.startswith('loro') or\
                                                                           i.startswith('nan') or\
                                                                    i.startswith('dell') \
                                                                    or i.startswith('sull')\
                                                                     or i.startswith('degl')
                                                                     or i.startswith('dall')
                                                                     or i.startswith('all')
                                                                          )])
info_data.head()
```

```python
info_data.to_csv('info_data.csv')
```
